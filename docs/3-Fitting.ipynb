Here we are going to use `scipy.optimize.curve_fit` to fit points to pre-defined curve function. This is not the only method for curve fitting in Python, but I cosider it to be very useful.

Details https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html#scipy.optimize.curve_fit

```python
# import packages
import numpy as np
import random
import pylab as plt
from scipy.optimize import curve_fit
```

### Fake data

Let's generate a random gaussian distribution of 1000 numbers as the data that we want to fit.

```python
# guass distribution 10000 points
mean = 10
sigma = 3
gauss_dist = [random.gauss(mean, sigma) for _ in range(10000)]

# distribution to 50 histogram points
y_data, x_data = np.histogram(gauss_dist, bins=50) 
x_data = x_data[1:]

# plot it
plt.plot(x_data, y_data, "x");
```

### Fitting

Let's define a python function (see [here](1-Basics.ipynb#Functions)) that represent a gaussian curve. 

```python
def gaussian(x, amplitude, mean, sigma):
    """ Define Gaussian curve function.
    """
    return amplitude * np.exp(-0.5*((x - mean) / 1/ sigma)**2)
```

`curve_fit` takes the defined curve function, x_data and y_data as arguments and return a list of optimized parameters (popt). 

```python
popt, _ = curve_fit(gaussian, x_data, y_data)

print('amplitude: ', round(popt[0]), ' mean: ', round(popt[1],1), ' sigma: ', round(abs(popt[2]), 1))
```

```python
# plot it
plt.plot(x_data, y_data, "x");
plt.plot(x_data, gaussian(x_data, *popt));
```

Although this example is for a gaussian fit, one can easily extent for other curves be editing the pre-defined curve function. Try it!
